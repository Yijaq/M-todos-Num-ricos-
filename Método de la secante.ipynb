```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Apr  9 13:50:08 2021

@author: Yilber Quinto
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def secante(f, x0, x1, tol, M):
    """
    

    Parameters
    ----------
    f :  Función 
         Función de números reales 
    x0 : float
         Punto inicial
    x1 : float
         Punto inicial 
    tol : float
         Tolerancia 
    M : int 
        Número máximo de interaciones 

    Returns
    -------
    Exito "El cero de la funcion es x = p"
    
    Fracaso " El método fracasó después de M interaciones "

    """
    
    valores = [x0, x1]
    error   = [np.nan, np.nan]
    
    i = 0
   
    while i < M: 
        x = (x0*f(x1) - x1*f(x0))/(f(x1) - f(x0))
        e = np.abs((x - x1)/x)
        
        valores.append(float(f'{x:.9f}'))
        error.append(float(f'{e:.9f}'))
        
        if e < tol:
            df = pd.DataFrame({'x_n':valores,'e_n':error})
            df.to_excel('Tabla_método_secante.xlsx') 
            display(pd.read_excel("Tabla_método_secante.xlsx",index_col=0))
            print('')
            print(f'El cero de la función es: {x:.9f}')
            return x
        else:
            x0 = x1
            x1 = x
        i = i + 1
    if i == M:
        df = pd.DataFrame({'x_n':valores,'e_n':error})
        df.to_excel('Tabla_método_secante.xlsx') 
        display(pd.read_excel("Tabla_método_secante.xlsx",index_col=0))
        print(f'No se logró una aproximacón despues de {M} iteraciónes')
```
